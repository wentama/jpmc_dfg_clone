```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
state = ['Arkansas', 'Oregon', 'California', 'Neveda', 'Washington']
```

```python
year = np.arange(2015,2021)

for i in year:
    yr = str(i)
    homeless_pop = pd.read_excel('data/2007-2021-PIT-Counts-by-State.xlsx', sheet_name = yr)
    
    # define attribute title in this loop
    tot_count = 'Overall Homeless, ' + yr
    unshel_indiv_18 = 'Unsheltered Homeless Individuals - Under 18, ' + yr
    unshel_indiv_18_24 = 'Unsheltered Homeless Individuals - Age 18 to 24, ' + yr
    homeless_pop['unshel_indiv_u24'] = homeless_pop[unshel_indiv_18] + homeless_pop[unshel_indiv_18_24]
    
    filter_df = homeless_pop[['State', tot_count, 'unshel_indiv_u24']]
    filter_df = filter_df.loc[filter_df['State'] != 'AS']
    filter_df = filter_df.loc[filter_df['State'] != 'MP']
    
    filter_df['unsheltered_ratio'] = filter_df['unshel_indiv_u24'].astype(int).astype(float) / filter_df[tot_count].astype(int).astype(float)
    
    sort_df = filter_df.sort_values('unsheltered_ratio', ascending = False)
    final_df = sort_df[['State', 'unsheltered_ratio']]
    final_df.rename(columns={'unsheltered_ratio': 'Unsheltered Ratio Under 24' } ,inplace=True)
    
    print("Year: " + yr)
    print(final_df.head(7))
    print('_____________________')
    
    
```
