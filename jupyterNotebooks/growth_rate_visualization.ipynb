```python
# James Pineda
# Intake and clean data file
import pandas as pd

def readSNAP(month, year):
    df = pd.read_excel(f"data/{month} {year}.xls", skiprows=3)
    #df = pd.read_excel(file, skiprows=3)
    # drop columns
    df = df.dropna(axis=1, how="all")
    # drop NaN rows
    df = df.dropna()
    df = df[['Substate/Region', 'Calc: SNAP Total PA and Non-PA People']]
    return df
```

```python
# Create list containing the dataframe in each year
dfListJAN = []
dfListJUL = []

for yr in range(2007, 2022):
    dfListJAN.append(readSNAP('jan', yr))
    dfListJUL.append(readSNAP('jul', yr))
```

```python
# January 2007
df = dfListJAN[0]

# Convert rest of list and add to dataframe
for i in range(1, len(dfListJAN)):
    df = pd.merge(df, dfListJAN[i], on="Substate/Region", how="outer")
    cName = str(i+2007)

df = df.set_axis(['Region', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015','2016', '2017', '2018', '2019', '2020', '2021'], axis=1, inplace=False)
df
```

```python
df['State'] = df['Region'].str.split(' ', expand=True)[1]
df['Region'] = df['Region'].str.split(' ', expand=True)[0]
df
```

```python
# 39061 => Hamilton
# 39049 => Franklin
# 39095 => Lucas

df = df[df["Region"].str.contains("39095")]
df = df.transpose()
df.drop(axis=1,)
# df.to_excel("LucasSNAP.xlsx", index=False)
```
