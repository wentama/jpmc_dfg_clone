```python
import pandas as pd
import numpy as np
```

```python
def readSheet(year):
    dfCounty = pd.read_excel("2007-2021-HIC-Counts-by-CoC.xls", sheet_name = str(year), header = 1)
    dfCounty = dfCounty.dropna()
    OH = dfCounty.loc[dfCounty['CoC Number'].str.contains('OH')]
    res = OH[['CoC Number', 'Total Year-Round Beds (ES, TH, SH)', 'Total Year-Round Beds (ES)']].copy()
    return res
```

```python
ex = readSheet(2021)
ex.sort_values(by='Total Year-Round Beds (ES, TH, SH)', ascending = True)
```

```python
def names(year):
    dfCounty = pd.read_excel("2007-2021-PIT-Counts-by-CoC.xlsx", sheet_name = str(year))
    dfCounty = dfCounty.dropna()
    OH = dfCounty.loc[dfCounty['CoC Number'].str.contains('OH')]
    OH = OH.loc[OH['CoC Name'] != 'Ohio Balance of State']
    resOH = OH[['CoC Name', 'CoC Number']]
    return resOH   

ex2 = names(2021)
ex2
```

```python
cols = ['CoC Number', 'CoC Name']
ex2 = ex2.merge(ex, how='left')
# ex2 = ex2.sort_values(by='Total Year-Round Beds (ES, TH, SH)', ascending = True)
```
