## Load Library

````python
import pandas as pd
````

## Ohio Beds

````python
beds = pd.read_excel("2007-2021-HIC-Counts-by-CoC.xls", header = 1)
beds.sample(3)
````

````python
beds = beds[["CoC Number", "Dedicated Youth Beds (ES, TH, SH)"]]
ohio_beds = beds["CoC Number"].str.contains("OH", case = False, na = False)
ohio_beds = beds[ohio_beds]
ohio_beds
````

## Ohio Homeless Population

````python
homeless_pop = pd.read_excel('2007-2021-PIT-Counts-by-CoC.xlsx', sheet_name = "2020")
homeless_pop.sample(3)
````

````python
ohio_filter = homeless_pop["CoC Number"].str.contains("OH", case = False, na = False)
ohio_homeless_pop = homeless_pop[ohio_filter]
ohio_homeless_pop
````

## Merge Ohio Beds and Ohio Homeless Population
- To include the ```CoC Name``` column for ```ohio_beds```

````python
ohio_beds = ohio_beds.merge(ohio_homeless_pop, how = "inner", on = ["CoC Number"])
ohio_final_beds = ohio_beds[["CoC Number", "CoC Name", "Dedicated Youth Beds (ES, TH, SH)"]]
ohio_final_beds
````

## Youth Homeless Ratio

````python
def Filter(string, substr):
    return [str for str in string if
             any(sub in str for sub in substr)]

Filter(list(ohio_homeless_pop.columns), "Overall")
````

````python
ohio_homeless_pop["youth_ratio"] = ohio_homeless_pop['Overall Homeless - Age 18 to 24, 2020'] / ohio_homeless_pop['Overall Homeless, 2020']
````

````python
ohio_homeless_pop[["CoC Name", "youth_ratio", "Overall Homeless, 2020"]]
````

## Find Bed Ratio 
- Availability of space 

````python
final_df = ohio_homeless_pop.merge(ohio_final_beds, how = "inner", on = ["CoC Number", "CoC Name"])
final_df 
````

````python
final_df["bed_ratio"] = final_df["Dedicated Youth Beds (ES, TH, SH)"] / final_df['Overall Homeless - Age 18 to 24, 2020']
````

````python
final_df["bed_pct"] = final_df["bed_ratio"] * 100
````

````python
final_df = final_df[["CoC Name", "bed_ratio", "bed_pct"]].sort_values("bed_pct").reset_index().drop(["index"], axis=1)
final_df
````

````python
# Toldeo, Hamilton, Franklin
final_counties = final_df.iloc[[2, 4, 5, ]]
final_counties
````

### Export DF

````python
final_counties.to_excel("counties_bed_space.xlsx")
````
