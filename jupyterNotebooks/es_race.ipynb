```python
def race_df(race):
    '''Create a DF per race'''
    race_df = ohio.loc[:,ohio.columns.str.contains(race)]
    race_df = race_df.loc[:, race_df.columns.str.contains("Under 25")]
    race_df["CoC Name"] = ohio["CoC Name"]
    race_df["es_unaccompanied_pct"] = race_df[f"Sheltered ES Homeless Unaccompanied Youth (Under 25) - {race}, 2021"]/ race_df[f"Sheltered Total Homeless Unaccompanied Youth (Under 25) - {race}, 2021"]
    race_df["es_parenting_pct"] = race_df[f"Sheltered ES Homeless Parenting Youth (Under 25) - {race}, 2021"]/ race_df[f"Sheltered Total Homeless Parenting Youth (Under 25) - {race}, 2021"]
    
    return race_df
        
#columns[[f"Sheltered ES Homeless Unaccompanied Youth (Under 25) - {} or African American, 2021"#
```

## Plots 

```python
plt.plot(black["CoC Name"], black["es_unaccompanied_pct"], label = "ES Unaccompanied")
plt.plot(black["CoC Name"], black["es_parenting_pct"], label = "ES Parenting")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0)
plt.xticks(rotation = 90)
plt.title("Percentage of Emergency Sheltered Black Youth")
```
